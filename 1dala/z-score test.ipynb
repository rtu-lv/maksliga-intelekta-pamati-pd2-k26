```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# load data
Raisin = np.dtype([
    ('Area', np.float32),
    ('MajorAxisLength', np.float32),
    ('MinorAxisLength', np.float32),
    ('Eccentricity', np.float32),
    ('ConvexArea', np.float32),
    ('Extent', np.float32),
    ('Perimeter', np.float32),
    ('Class', str),
])

df = pd.read_csv('../data.csv', dtype=Raisin)

color_map = {
    "Kecimen": "red",
    "Besni": "blue"
}

numeric_cols = df.select_dtypes(include=[np.number])

def plot_data(threshold=3.0):
    z_scores = (numeric_cols - numeric_cols.mean()) / numeric_cols.std()
    mask = ~(np.abs(z_scores) > threshold).any(axis=1)

    filtered = df[mask]

    plt.figure(figsize=(5, 4))
    plt.scatter(
        filtered["MajorAxisLength"],
        filtered["MinorAxisLength"],
        c=filtered["Class"].map(color_map),
        alpha=0.6
    )
    plt.title(f"Outlier threshold: {threshold}")
    plt.xlabel("MajorAxisLength")
    plt.ylabel("MinorAxisLength")
    plt.show()

slider = widgets.FloatSlider(
    value=3.0,
    min=1.0,
    max=6.7,
    step=0.1,
    description='Z threshold'
)

widgets.interactive(plot_data, threshold=slider)
```
